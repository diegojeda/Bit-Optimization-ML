## **2.2   Tipos de Variables**

 <img src="https://enciclopediaeconomica.com/wp-content/uploads/2018/12/variables.jpg" align="Right" height="350" width="420" />

Es importante definir los tipos de variables que podemos encontrar en nuestros datos, con el fin de dar el mejor y mas adecuado tratamiento a cada una.

**Variables Cualitativas:** Las variables cualitativas son aquellas caracter√≠sticas o cualidades que no pueden ser calculadas con n√∫meros, sino que son clasificadas con palabras.

*   *Cualitativa nominal:* aquellas variables que no siguen ning√∫n orden en espec√≠fico. Por ejemplo, los colores, tales como el negro, naranja o amarillo.
*   *Cualitativa ordinal:* aquellas que siguen un orden o jerarqu√≠a. Por ejemplo, el nivel socioecon√≥mico alto, medio o bajo.
*   *Cualitativa binaria:* variables que permiten tan solo dos resultados. Por ejemplo, s√≠ o no; hombre o mujer.

**Variables Cuantitativas:** Las variables cuantitativas son aquellas caracter√≠sticas o cualidades que s√≠ pueden expresarse o medirse a trav√©s de n√∫meros.

*   *Cuantitativa discreta:* aquella variable que utiliza valores enteros y no finitos. Por ejemplo, la cantidad de familiares que tiene una persona, tal como 2, 3, 4 o m√°s.
*   *Cuantitativa continua:* aquella variable que utiliza valores finitos y objetivos, y suele caracterizarse por utilizar valores decimales. Por ejemplo, el peso de una persona, tal como 64.3 kg, 72.3 kg, etc.

Para evaluar el rendimiento de las aproximaciones de la estimaci√≥n de *f* se utilizan m√©tricas que nos ayudan a medir "que tan bien aproxima *f* a mi valor real".

## **2.3   Paradigma Sesgo - Varianza**
Cuando deseamos generar un modelo, usamos los datos disponibles para que este "aprenda de ellos" y luego estime valores acerca de "nuevos datos".

En ese orden de ideas, los datos totales se deben dividir entre *Train* y *Test*. De esta manera, nuestros modelo lo generamos con nuestros datos de *Train* y los probamos con los datos de *Test*.

* Si el modelo predice muy bien los datos de test, no aprendi√≥ de los datos, "se aprendi√≥ los datos" (Alto Sesgo), cuando evalue el modelo con los datos de test, no los predecir√° bien (Alta Varianza).
* Un enfoque puede ser dejar que el modelo no prediga tan bien mis datos de test (Minimizar el Sesgo) para que cuando vea datos nuevos (test) prediga mejor (Minimizar la Varianza).
* Sin embargo, existe un un punto donde disminur el sesgo juega en contra y la varianza de mi modelo: al ver datos nuevos tiene un error mayor.

La clave es encontrar un punto de equilibrio que me permita minimizar el error cuando el modelo vea datos nuevos.

## **7.2  Modelos de Regresion Lineal**

El an√°lisis de Regresi√≥n es un subcampo de ‚Äúmachine learning supervisado‚Äù. Su prop√≥sito es establecer un modelo para la relaci√≥n entre un cierto n√∫mero de caracter√≠sticas y una variable objetivo continua.

En los problemas de regresi√≥n perseguimos obtener una respuesta cuantitativa, como por ejemplo, predicciones sobre precios de inmuebles o el n√∫mero de segundos que alguien dedicar√° a visualizar un v√≠deo.

Dado un conjunto de puntos, el algoritmo de regresi√≥n establecer√° un modelo para ajustar la relaci√≥n de dependencia entre una caracter√≠stica espec√≠fica independiente (un valor de la variable independiente x) y el valor ‚Äúresultado‚Äùcorrespondiente (un valor de la variable dependiente y).
Esta relaci√≥n se realiza estableciendo una l√≠nea arbitraria y calculando la distancia de la recta a los puntos de datos correspondientes a los valores (x,y). Esta distancia, las l√≠neas verticales, son los ‚Äúresiduos‚Äù o los ‚Äúerrores de predicci√≥n‚Äù.

El algoritmo de regresi√≥n recalcular√° (y mover√°) la recta con cada interacci√≥n, buscando aquella que mejor se ajuste a los puntos de datos (x,y), o en otras palabras, la l√≠nea con el menor error (la m√°s cercana al m√°ximo n√∫mero de puntos).

### **7.2.1  Regresion Lineal**

En los modelos de regresion lineal se busca obtener el valor de los coerficientes ($\beta_i$) que acompa√±an las variables predictoras, buscando minimizar *Las Diferencias de los Errores al Cuadrado*:


$$RSS = \sum_{i=1}^{n}{(y_i-f(x_i))^2} = \sum_{i=1}^{n}{(y_i-(\alpha + \beta x_i))^2}$$

Procederemos a escalar los datos, para que el modelo los pueda procesar, importaremos el modelo, dividir la base en:

- *Train*: que corresponden a las observaciones con las cuales el modelo "aprender√° de ellas".
- *Test*: que son las observaciones con las cuales se probar√° la efectividad.

Instanciaremos el modelo, lo entrenaremos (con la data de train), luego generaremos las predicciones (con la data de test).



```python
# Importamos librerias para escalar los datos, dividir el dataset y traer el modelo
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error , r2_score
import warnings
warnings.filterwarnings("ignore")

# Creamos las matrices para los modelos
X = base.drop('Log ROP', axis=1)
y = base[['Log ROP']]

# Escalamos los datos
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
pd.DataFrame(X).head();
```

```python
# Dividimos los datos de manera aleatoria con el 25% para test y 75% para train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)

# Creamos los modelos LINEAL, LASSO Y RIDGE
linear = linear_model.LinearRegression()

# Ajutstamos el modelo lineal
linear.fit(X_train, y_train)

# Diccionario con el modelo
modelos = [("Regresion Lineal",linear)];
```

Ahora, prodeceremos a mostar los resultados:

```python
def tabla_result (modelos,X_test,y_test):
    '''
    modelos: Diccionario con los modelos a evaluar
    X_test : Matriz X con los valores de test
    y_test : Matriz y con los valores de test
    '''
    y_pred = []
    MSE = []
    MSE_lin = []
    R2 = []
    R2_lin = []
    
    for m_name,m in modelos:
        y=m.predict(X_test)
        y_pred.append(y)
        
        MSE.append(mean_squared_error(y_test,y))
        MSE_lin.append(mean_squared_error(np.exp(y_test),np.exp(y)))
        
        R2.append(r2_score(y_test,y))
        R2_lin.append(r2_score(np.exp(y_test),np.exp(y)))
        
    return (y_pred, MSE, R2, R2_lin, MSE_lin)
```

```python
# Extraemos los resultados
y_pred, MSE, R2, R2_lin, MSE_lin = tabla_result(modelos,X_test,y_test)
```

```python
# Presentamos los datos en una tabla
Res_Lin = {"Modelo":["Linear"],
           "MSE":MSE,
           "R2":R2}
Res_Lin = pd.DataFrame(Res_Lin)
Res_Lin.set_index("Modelo",inplace=True)
Res_Lin

# Mostramos los coeficientes de la regresion
a=pd.DataFrame(base.drop('Log ROP', axis=1).columns)
b=pd.DataFrame(linear.coef_)
Coef_Lin=pd.concat([a,b.T], axis=1)
Coef_Lin.columns = ['Predictor','Coeficientes Lineal']
Coef_Lin.set_index("Predictor",inplace=True)
Coef_Lin
```

### **7.2.2  Regresion Lineal Con Penalizacion**

En los modelos de regresion lineal con penalizacion se busca obtener el valor de los coerficientes ($\beta_i$) que acompa√±an las variables predictoras, buscando minimizar Las Diferencias de los Errores al Cuadrado mas  un *Coeficiente de Penalizacion*:


 - Regresion Lineal: $$RSS = \sum_{i=1}^{n}{(y_i-f(x_i))^2} = \sum_{i=1}^{n}{(y_i-(\alpha + \beta x_i))^2}$$

  - Regresion Ridge: $$ \sum_{i=1}^{n}{(y_i-(\alpha + \beta x_i))^2} + \lambda \sum_{j=1}^{p}{\beta_i^2}$$

   - Regresion Lasso: $$ \sum_{i=1}^{n}{(y_i-(\alpha + \beta x_i))^2} + \lambda  \sum_{j=1}^{p}{\lvert \beta_i \lvert}$$

```python
# Importarmos librerias para hacer cross-validation para hallar los parametros optimos de Aplha
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

# Establecemos los parametros a usar para variar el parametro de penalizazion Alpha
parameters = {'alpha': np.concatenate((np.arange(0.01,1,0.001), np.arange(1, 5, 0.5), np.arange(5, 50, 1)))}

# Creamos los modelos LASSO Y RIDGE
lasso = linear_model.Lasso()
ridge = linear_model.Ridge()

# Generamos los diferentes modelos para diferentes valores Alpha
gridlasso = GridSearchCV(lasso, parameters, scoring ='r2')
gridridge = GridSearchCV(ridge, parameters, scoring ='r2')

# Ajustamos los modelos con los parametros a variar
gridlasso.fit(X_train, y_train)
gridridge.fit(X_train, y_train)

# Diccionario con el modelo
modelos = [("Regresion Lineal",linear),
           ("Regresion Ridge",gridridge),
           ("Regresion Lasso",gridlasso)
          ]

# Extraemos los resultados
y_pred, MSE, R2, R2_lin, MSE_lin = tabla_result(modelos,X_test,y_test)
```

```python
# Mostramos los valores de las metricas de cada uno de los modelos:
ind = ["Lineal","Ridge","Lasso"]
Alphas = [ "NA",gridridge.best_params_.get("alpha") , gridlasso.best_params_.get("alpha")]

Best_Res_Lin = pd.DataFrame({"Modelo":ind,"Alpha Optimo":Alphas , "R2 Score": R2, "MSE":MSE})
Best_Res_Lin.set_index("Modelo")

# Mostramos los valores de los coeficientes de la regresion con los mejores valores de penalizacion
a=pd.DataFrame(base.drop('Log ROP', axis=1).columns)
b=pd.DataFrame(gridlasso.best_estimator_.coef_)
c=pd.DataFrame(gridridge.best_estimator_.coef_)
d=pd.DataFrame(linear.coef_)
Res=pd.concat([a,d.T,c.T,b], axis=1)
Res.columns = ['Predictor','Coeficientes Lineal',"Coeficientes Ridge","Coeficientes Lasso"]
Res.set_index("Predictor")
```

```python
# updatable plot
# a minimal example (sort of)

class PlotLosses(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
        plt.legend()
        plt.show();
        
plot_losses = PlotLosses()
```
